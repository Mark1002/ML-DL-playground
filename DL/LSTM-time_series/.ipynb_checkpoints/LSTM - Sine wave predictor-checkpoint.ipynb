```python
import pandas as pd
import numpy as np
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.layers import TimeDistributed
from keras.models import Sequential
import time
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
series = pd.read_csv('sine-wave.csv', header=None )
```

```python
series.head()
```

```python
series.shape
```

```python
pyplot.figure(figsize=(20,6))
pyplot.plot(series.values)
pyplot.show()
```

```python
pyplot.figure(figsize=(20,6))
pyplot.plot(series.values[:50])
pyplot.show()
```

```python
window_size = 50

series_s = series.copy()
for i in range(window_size):
    series = pd.concat([series, series_s.shift(-(i+1))], axis = 1)
    
series.dropna(axis=0, inplace=True)
```

```python
series.head()
```

```python
series.shape
```

```python
nrow = round(0.8*series.shape[0])
train = series.iloc[:nrow, :]
test = series.iloc[nrow:,:]
```

```python
from sklearn.utils import shuffle
train = shuffle(train)
```

```python
train_X = train.iloc[:,:-1]
train_y = train.iloc[:,-1]
test_X = test.iloc[:,:-1]
test_y = test.iloc[:,-1]
```

```python
train_X = train_X.values
train_y = train_y.values
test_X = test_X.values
test_y = test_y.values
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
train_X = train_X.reshape(train_X.shape[0],train_X.shape[1],1)
test_X = test_X.reshape(test_X.shape[0],test_X.shape[1],1)
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
# Define the LSTM model
model = Sequential()
model.add(LSTM(10, input_shape=(None,1)))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")
model.summary()
```

```python
start = time.time()
model.fit(train_X, train_y, batch_size=128, epochs=20, validation_split=0.1)
print("> Compilation Time : ", time.time() - start)
```

```python
# Doing a prediction on all the test data at once
preds = model.predict(test_X)
```

```python
preds
```

```python
actuals = test_y
```

```python
mean_squared_error(actuals,preds)
```
